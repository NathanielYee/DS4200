### Part 3 - Basic data analysis with Altair (40 points)

##### Question 1 (10 points)

Recreate the figure in Part 3 Question 2 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and weather type.

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import altair as alt
import matplotlib.pyplot as plt
from vega_datasets import data

weather = data.seattle_weather()
weather
```

```python
alt.Chart(weather).mark_boxplot().encode(
    alt.X('temp_max:Q',title='Maximum Temperature'),
    alt.Y('weather:N',title='Weather'),
    alt.Color('weather')
).properties(
    width = 500,
    height = 500,
    title = 'Maximum Temperature vs. Weather'
)
```

##### Question 2 (10 points)

Recreate the figure in Part 3 Question 3 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type.

```python
alt.Chart(weather).mark_point().encode(
    alt.X('temp_min',title='Minimum Temperature'),
    alt.Y('temp_max',title='Maximum Temperature'),
    alt.Color('weather')
).properties(
    width = 900,
    height = 600,
    title = 'Minimum vs. Maximum Temperature'
)

```

##### Question 3 (10 points)

Create a facet figure in Altair with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python
# Ask about this facet plot and why it is not working?
alt.Chart(weather).mark_point().encode(
    alt.X('temp_min:Q',title='Minimum Temperature'),
    alt.Y('temp_max:Q',title='Maxiumum Temperature'),
    alt.Color('wind:Q',scale=alt.Scale(scheme='category10')),
    alt.Tooltip(['temp_min:Q','temp_max:Q','wind:Q'])
).facet(facet='weather:N',title='Weather',columns=3).interactive()
```

##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Note, just use the `date` as X attribute, no need to aggragate for years and you will get a wide figure. Include necessary title and labels.

```python

max_temp = alt.Chart(weather).mark_point().encode(
    alt.X('date:N',title='Date'),
    alt.Y('mean(temp_max):Q',title='Average Temp'),
    alt.Color('weather:N')

)
min_temp = alt.Chart(weather).mark_point(shape="diamond").encode(
    alt.X('date:N',title='Date'),
    alt.Y('mean(temp_min):Q',title='Average Temp'),
    alt.Color('weather:N')

)
(max_temp + min_temp).properties(
    title= 'Average Min and Max Temps By Date'


)
```

##### Question 1 (5 points)

Following is a density plot for the wind speed for all five types of weather. Is there any problems in the figure? Try to name at least two ways you can do to re-design the figure (change the mark/channel/encoding etc.)

### Answer
- One way you could redesign the figure is using a different mark instead of using a density plot using either a line plot or mark plot since the density seems to overshadow other points towards the back
-The other way you could redesign the figure is working with facet plots to increase the seperation and making each weather type into its own plot so we can fully visualilze each of the data points 

```python
# Create density plots using Altair
density_plot = alt.Chart(weather).transform_density(
    'wind',
    groupby=['weather'],
    as_=['wind', 'density']
).mark_area(opacity=0.7).encode(
    alt.X('wind:Q'),
    alt.Y('density:Q'),
    alt.Color('weather:N', scale=alt.Scale(
            range=['lightgray', 'palegreen', 'skyblue', 'lightcoral', 'crimson']))
).properties(
    title='Density Plot of Wind for Sun, Drizzle, and Snow'
)

# Show the plot
density_plot.display()
```
