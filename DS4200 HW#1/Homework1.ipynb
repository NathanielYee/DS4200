### Part 3 - Basic data visualization with Python (50 points)

Here is a dataset of Seattle weather in 4 years. Load the data and generate corresponding data visualizations. Make sure each figure has title, labels and legends if necessary. Add one or two sentences in each figure to talk about what you have learned from the figure. 

```python
import pandas as pd
from scipy.cluster.hierarchy import average

weather = 'weather.csv'
weather = pd.read_csv(weather)
weather.columns = weather.columns.map(str)
#weather = pd.DataFrame(weather)
weather
```

##### Question 1 (10 pts)

Generate a data visualization to show the relationship between precipitation and max temperature.  

```python
import seaborn as sns
import matplotlib.pyplot as plt
# This plot taught me that as that there is a negative correlation between max temperature and amount of precipitation as seen by the correlation line from the regression.
# use regplot to 
plt.figure(figsize=(20,10))
sns.regplot(x=weather['precipitation'],y=weather['temp_max'])
plt.title('Relationship Between Precipitation and Max Temperature')
plt.xlabel('Precipitation')
plt.ylabel('Max Temperature')
```

##### Question 2 (10 pts)

Generate a data visualization to show the relationship between max temperature and weather type.  

```python
# use box plot to see temperature ranges for max temp by type of weather
# This plot reveals that "sun" weather is associated with the highest range of maximum temperatures, typically between 15Â°C and 25Â°C, with some outliers reaching up to 35Â°C. Additionally, "snow" weather occurs at much lower maximum temperatures, generally below 10Â°C.
sns.boxplot(weather,x='temp_max',y='weather',hue='weather')

# adjust labels and title 
plt.title('Relationship Between Max Temperature and Weather Type')
plt.xlabel('Max Temperature')
plt.ylabel('Weather Type')
```

##### Question 3 (10 pts)

Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type. 

```python
# Create scatterplot with hue as the parameter to differentiate the weather types
# This scatter plot shows a strong positive correlation between minimum and maximum temperatures, regardless of the weather type. It also suggests that "sun" weather (green dots) tends to occur at higher minimum and maximum temperatures compared to other weather types, while "snow" (red dots) is limited to the lowest temperature range.
plt.figure(figsize=(10,6))
sns.scatterplot(weather,x='temp_min',y='temp_max',hue='weather')

#add labels for ease of reading
plt.title('Relationship Between Minimum and Maximum Temperature')
plt.xlabel('Min Temperature')
plt.ylabel('Max Temperature')
```

##### Question 4 (10 pts) 

Generate a data visualization to show the relationship between max temperature, min temperature and wind speed. 

```python
# This scatter plot also demonstrates a positive correlation between minimum and maximum temperatures. Additionally, the size of the points represents wind speed, with larger points indicating higher wind speeds. The plot suggests that higher temperatures (both minimum and maximum) tend to coincide with stronger winds, as indicated by larger dots appearing more frequently at higher temperatures.
# Adjust fig size to fit axis
plt.figure(figsize=(20,10))
# Use reg plot to see the relationshp between min/max temp and windspeed
sns.scatterplot(data=weather,x='temp_min',y='temp_max',size='wind',color ='blue')
#sns.regplot(data=weather,x='wind',y='temp_max',label='Max Temp',color='red')

#create titles and labels as well as display legend
plt.title('Relationship Between Min and Max Temps And Windspeed')
plt.legend()
plt.xlabel('Minimum Temperature')
plt.ylabel('Maximum Temperature')
```

##### Question 5 (10 pts)

Calculate the average temperature for each day (average on min and max). Plot the average temperature with the date.

```python
# This plot taught me that the average temperature is cyclical and may be a result of the changing seasons as time goes on. 
import matplotlib.dates as mdates
# use pandas to create a new column with average temps per day
weather['avg_temp'] = (weather['temp_min'] + weather['temp_max']) / 2
# weather

# Adjust Fig Size and then use scatterplot to show temp averages for the dates
plt.figure(figsize=(30,6))
sns.scatterplot(data=weather,x=weather['date'],y=weather['avg_temp'])

# Add titles and format the x axis to make it more readable 
plt.title('Average Temperature By Date')
plt.xlabel('Date')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

plt.ylabel('Average Temperature')
```
