```python
import seaborn as sns
import altair as alt
import matplotlib.pyplot as plt
import pandas as pd
from scipy.special import title
```

```python
titanic_data = sns.load_dataset('titanic')
print(titanic_data.head())
```

```python
# Plot 1 Bar Plot
df_class_alive = titanic_data['class'].value_counts().reset_index()
df_class_alive.columns = ['class','alive']
alt.Chart(df_class_alive).mark_bar().encode(
    alt.X('class:N',title='Class'),
    alt.Y('alive:Q',title='Survived'),
    color = 'class'
).properties(
    title = 'Number of Survivals By Passenger Class'
)
```

```python
  alt.Chart(titanic_data).mark_point().encode(
    alt.X('age:Q',title='Passenger Age'),
    alt.Y('fare:Q',title='Fare Price'),
    
    color = 'class'
).properties(
    title = 'Age vs. Fare Price' )
```

```python
alt.Chart(titanic_data).mark_bar().encode(
    alt.X('age:Q',bin=alt.Bin(maxbins=10),title="Age"),
    alt.Y('count(survived)',title='Survived'),
    alt.Color('sex:N')
).properties(
    title = 'Age and Survival Rates By Sex'
)
```
