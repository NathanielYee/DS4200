```python
import pandas as pd
import seaborn as sns
import yfinance as yf
import matplotlib.pyplot as plt 


spy = yf.Ticker('SPY')
data = yf.download('SPY',period='1y')
data['30_day_ma'] = data['Adj Close'].rolling(window=30).mean()
print(data)

# Create the Plot
sns.lineplot(data,x='Date',y='Adj Close',label='Adj Close')
sns.lineplot(data,x='Date',y='30_day_ma',label='30 Day MA')

plt.title('SPY Historical Adjusted Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
```
